```python
#5.9.2023
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO

# Define the length of the genome sequence in megabytes
desired_size_mb = 5  # Change this to your desired size

# Calculate the length in base pairs!! (1 megabyte is approximately 1 million base pairs)
length_bp = desired_size_mb * 5**6 #10^6 = 1,000,000

# Generate a random DNA sequence of the specified length
import random
random.seed(42)  # Set a random seed for reproducibility, 42 is a random number just to store seed info. it could have been either 35 or 99
random_sequence = ''.join(random.choice('ACGT') for _ in range(length_bp))
#to initialize the random number generator. The random seed is an initial value that determines the sequence of random numbers generated by the random number generator.

# Create a Biopython Seq object from the random sequence
random_seq = Seq(random_sequence)

# Create a SeqRecord to store the sequence with a description
seq_record = SeqRecord(random_seq, id="RandomGenome", description=f"Random genome of {desired_size_mb} MB")

# Save the sequence to a FASTA file
output_file = "random_genome.fasta"
SeqIO.write(seq_record, output_file, "fasta")

print(f"Random genome sequence of {desired_size_mb} MB generated and saved to {output_file}") #F-strings in Python provide a more readable, more concise and efficient way to format strings with embedded variables
```
